# CoinMarketCap API
- In order to manipulate the DataFrames below, you'll need to sign up for your API Key at this link:
 https://pro.coinmarketcap.com/signup/

```python
your_key_here = input('Enter your unique API Key: ')
```

```python
from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json
import pandas as pd

headers ={'Accepts':'application/json','X-CMC_PRO_API_KEY':your_key_here,}

url ='https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
parameters = {'start':'1','limit':'5000','convert':'USD'}

session = Session()
session.headers.update(headers)

try:
  response = session.get(url, params=parameters)
  data_jason = json.loads(response.text)
except (ConnectionError, Timeout, TooManyRedirects) as e:
  print(e)
```

### Cryptocurrency Prices sorted by Market Cap

```python
df = pd.DataFrame(data_jason['data'])
df[['cmc_rank','symbol','name','max_supply','circulating_supply','quote']][0:10] #top 10
```
